```python
from Bees import Bees

# import sys  
# !{sys.executable} -m pip install --user matplotlib
import matplotlib.pyplot as plt 


size_x = 10
size_y = 10
best_site = 2
elite_site = 4
nre = 4
nrb = 2
ns = 1

bee_algo = Bees(size_x, size_y, best_site, elite_site, nre, nrb, ns)

solutions = bee_algo.solutions

figure, axes = plt.subplots(figsize=(10, 10)) 

for i in range(0, size_x * size_y):
    row = int(i / size_y)
    col = i % size_x
    cc = plt.Circle(( col/10 + .05 , row/10 + .05), 0.04, color="green" )
    axes.add_artist(cc)
    plt.text(col/10 + .03, row/10 + .03, solutions[i])
ar = plt.Arrow(.01, .02, .07, .05, width=.01)  
axes.add_artist(ar)

axes.set_aspect( 1 )
plt.title( 'Field' ) 
plt.show()
```

```python

bee_algo.do_first_step()
scouts = bee_algo.scout_bees
print(scouts)
flower_patches = bee_algo.flower_patches
print(flower_patches)
solutions = bee_algo.solutions

figure, axes = plt.subplots() 
for i in range(0, size_x * size_y):
    row = int(i / size_y)
    col = i % size_x
    cc = plt.Circle(( col/10 + .05 , row/10 + .05), 0.04, color="green" )
    if i in scouts:
        cc = plt.Circle(( col/10 + .05 , row/10 + .05), 0.04, color="blue" )
    axes.add_artist(cc)
    plt.text(col/10 + .03, row/10 + .03, solutions[i])
ar = plt.Arrow(.01, .02, .07, .05, width=.01)  
axes.add_artist(ar)

for patch in flower_patches:
    row = int(patch[0] / size_y)
    col = patch[0] % size_x
    rect = plt.Rectangle((col/10-patch[1]/10, row/10-patch[1]/10), patch[1]/10*2+.1, patch[1]/10*2+.1, fill=False)
    axes.add_artist(rect)

axes.set_aspect( 1 )
plt.title( 'Field with scouts(Blue) and the flower patches' ) 
plt.show()

```

```python

bee_algo.do_second_step()
foragers = bee_algo.foragers
print(foragers)
print(scouts)
f = []
for x in foragers:
    for y in x[1:]:
        f.append(y)
print(f)

figure, axes = plt.subplots() 
for i in range(0, size_x * size_y):
    row = int(i / size_y)
    col = i % size_x
    cc = plt.Circle(( col/10 + .05 , row/10 + .05), 0.04, color="green" )
    if i in scouts:
        cc = plt.Circle(( col/10 + .05 , row/10 + .05), 0.04, color="blue" )
    if i in f:
        cc = plt.Circle(( col/10 + .05 , row/10 + .05), 0.04, color="red" )
    axes.add_artist(cc)
    plt.text(col/10 + .03, row/10 + .03, solutions[i])
ar = plt.Arrow(.01, .02, .07, .05, width=.01)  
axes.add_artist(ar)

for patch in flower_patches:
    row = int(patch[0] / size_y)
    col = patch[0] % size_x
    rect = plt.Rectangle((col/10-patch[1]/10, row/10-patch[1]/10), patch[1]/10*2, patch[1]/10*2, fill=False)
    axes.add_artist(rect)

axes.set_aspect( 1 )
plt.title( 'Field with scouts(Blue), foragers(Red), and the flower patches' ) 
plt.show()
```
